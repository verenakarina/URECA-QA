```python
excel_file = '{}_final.xlsx'.format(dataset_name)

set_type = ['train', 'dev', 'test']
dataset = ['reader', 'retrieval']

for i in dataset:
    for j in set_type:
        xlsx_data = pd.read_excel(excel_file, sheet_name = j)
        data = []
        for line in xlsx_data.iterrows():
            id = line[1][1]
            title = line[1][2]
            qn = line[1][3]
            corpus = line[1][4]
            ans	= line[1][5]
            positive_ctxs = line[1][6]
            negative_ctxs = line[1][7]    
            positive_title = line[1][8]
            negative_title = line[1][9]

            if i == 'retrieval':     

                results_item = {
                                "question": qn,
                                "answers": [ ans ],
                                "positive_ctxs": [
                                    {
                                        "title": positive_title,
                                        "text": positive_ctxs,
                                    }    
                                ],
                                "negative_ctxs": [
                                    {                               
                                        "title": negative_title,
                                        "text": negative_ctxs,
                                    }    
                                ],
                            }
                
            if i == 'reader':
                results_item = {
                                "question": qn,
                                "answers": [ ans ],
                                "ctxs": [
                                    {
                                        "id": str(id),                                
                                        "title": title,
                                        "text": corpus,
                                        "score": "103.232",
                                        "has_answer": True,
                                    }
                                ],
                            }                    
                
                
            data.append(results_item)
            with open("{}-{}_{}.json".format(dataset_name,i,j), 'w') as f:
                f.write(json.dumps(data, indent = 4) + "\n")
```
