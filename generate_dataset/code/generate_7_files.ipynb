```python
dataset_name = "COVID"
dataset = "/Users/verenakarina/Desktop/URECA/generate_dataset(updated)/covid_dataset_revised.csv"
csv_data = pd.read_csv(dataset)
df = pd.DataFrame(csv_data)
df
```

```python
#create index column
index = [i for i in range(df.shape[0])]
df['index'] = index
columns = ["index", "question", "answer", "label_len", "positive_title", "positive_label", "negative_title", "negative_label"]
df = df[columns]
df
```

```python
features = ["question", "answer", "label_len", "positive_title", "positive_label", "negative_title", "negative_label"]

X = df.loc[:, features]
y = df.loc[:, "index"]

X_train, X_dev_test, y_train, y_dev_test = train_test_split(X, y, random_state=0, train_size = .8)
X_dev, X_test, y_dev, y_test = train_test_split(X_dev_test, y_dev_test, random_state=0, train_size = .5)

dev = pd.concat([y_dev, X_dev], ignore_index=False, axis = 1)
test = pd.concat([y_test, X_test], ignore_index=False, axis = 1)
train = pd.concat([y_train, X_train], ignore_index=False, axis = 1)

with pd.ExcelWriter('{}_final.xlsx'.format(dataset_name)) as writer:
    df.to_excel(writer, sheet_name='all')  
    train.to_excel(writer, sheet_name='train')
    test.to_excel(writer, sheet_name='test')
    dev.to_excel(writer, sheet_name='dev')
```

```python
excel_file = '{}_final.xlsx'.format(dataset_name)
xlsx_data = pd.read_excel(excel_file, sheet_name = 'test')
df_test = pd.DataFrame(xlsx_data)
features = ["index", "question", "answer", "label_len", "positive_title", "positive_label", "negative_title", "negative_label"]
df_test = df_test[features]
df_test.to_csv('{}_test.csv'.format(dataset_name))
```

```python
excel_file = '{}_final.xlsx'.format(dataset_name)

set_type = ['train', 'dev', 'test']
dataset = ['reader', 'retrieval']

for i in dataset:
    for j in set_type:
        xlsx_data = pd.read_excel(excel_file, sheet_name = j)
        data = []
        for line in xlsx_data.iterrows():
            id = line[1][1]
            qn = line[1][2]
            ans = line[1][3]
            label_len = line[1][4]
            title = line[1][5]
            corpus = line[1][6]
            positive_title = line[1][5]
            positive_ctxs = line[1][6]
            negative_title = line[1][7]    
            negative_ctxs = line[1][8]

            if i == 'retrieval':     

                results_item = {
                                "question": qn,
                                "answers": [ ans ],
                                "positive_ctxs": [
                                    {
                                        "title": positive_title,
                                        "text": positive_ctxs,
                                    }    
                                ],
                                "negative_ctxs": [
                                    {                               
                                        "title": negative_title,
                                        "text": negative_ctxs,
                                    }    
                                ],
                            }
                
            if i == 'reader':
                results_item = {
                                "question": qn,
                                "answers": [ ans ],
                                "ctxs": [
                                    {
                                        "id": str(id),                                
                                        "title": title,
                                        "text": corpus,
                                        "score": "103.232",
                                        "has_answer": True,
                                    }
                                ],
                            }                    
                
                
            data.append(results_item)
            with open("{}-{}_{}.json".format(dataset_name,i,j), 'w') as f:
                f.write(json.dumps(data, indent = 4) + "\n")
```
