```python
import pandas as pd

#open file csv
input_file = "../input/initial-data-qa/5k_new.csv"
df = pd.read_csv(input_file, delimiter = "\t",
                 names = ['id_par','title','text','q','a','annotators'], header = None, encoding = 'utf-8')

df.head(10)
```

```python
import torch
torch.cuda.empty_cache()
from transformers import pipeline
```

```python
#DEFINE LANGUAGES
language1 = "en"
intermediate="jap"
language2 = "rw"
language3 = "ru"
file = "retrieval"
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
model = AutoModelForSeq2SeqLM.from_pretrained(f"Helsinki-NLP/opus-mt-{language1}-{language2}")
tokenizer = AutoTokenizer.from_pretrained(f"Helsinki-NLP/opus-mt-{language1}-{language2}")
```

```python
model.to('cuda:0')
pipe = pipeline("translation", model=model, tokenizer=tokenizer, device = 0)
```

```python
from transformers import pipeline
from tqdm import tqdm

#FIRST TRANSLATION
first = []
for question in tqdm(df['q']):
    translated_str = pipe(question)[0]['translation_text']
    first.append(translated_str)
```
